### creating a NumPy

```python
import numpy as np
```

```python
nums= np.array([1,3,5,7,9])
print("array:",nums)
```

### add 2 to each element

```python
nums_2=nums + 2
print("add_array:",nums_2)
```

### multiply each element by 3

```python
nums_3=nums * 3
print("multiplied_array:",nums_3)
```

### find and print sum

```python
nums =np.sum(nums)
print(nums)
```

### find and print the max value

```python
nums=np.max(nums)
print(nums)
```

### find and print mean

```python
nums=np.mean(nums)
print(nums)
```
