```python
import numpy as np
```

```python
nums = np.array([1, 3, 5, 7, 9])
num_2 = nums + 2
print(f"Array after adding 2: {num_2}")
```

```python
num_3 = nums * 3
print(f"Array after multiplying by 3: {num_3}")
```

```python
array_sum = np.sum(nums)
array_max = np.max(nums)
array_mean = np.mean(nums)
```

```python
print(f"Sum: {array_sum}")
print(f"Maximum Value: {array_max}")
print(f"Mean: {array_mean}")
```
