create a NumPy array called nums containing the values 1, 3, 5, 7, 9, print the array, add 2 to each element and store the result in a new array, multiply each element by 3, and then use NumPy methods to find and print the sum, maximum value, and mean of the original array.

```python
import numpy as np
```

```python
nums = np.array([ 1, 3, 5, 7, 9])
print(nums)
```

```python
new_array = nums + 2
print(new_array)
```

```python
array_3 = nums * 3
print(array_3)
```

```python
total_array = np.sum(nums)
print(total_array)
max_array = np.max(nums)
print(max_array)
mean_array = np.mean(nums)
print(mean_array)
```
